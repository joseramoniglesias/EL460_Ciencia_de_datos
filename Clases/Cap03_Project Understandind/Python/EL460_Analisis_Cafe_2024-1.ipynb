```python
#CONSTRUCCION DEL MODELO PREDICTIVO UTILIZANDO EL METODO DE REGRESION LINEAL
```

```python
# Iniciamos el proceso para determinar el modelo de regresion lineal, de la analitica determinista
# Asignamos a nuestra variable de entrada X (En este caso corresponde al Area Nacional) para entrenamiento (train).
# Asignamos a la variable dependiente Y (En este caso corresponde a la Produccion Nacional) para entrenamiento (train).
dataX =produccion_df[["Area Nacional (ha)"]]
X_train = np.array(dataX)
y_train = produccion_df['Produccion Nacional (ton)'].values
#print ('X_train',X_train)
#print ('y_train',y_train)
```

```python
# Creamos la función objeto para determinar la Regresión Lineal Y= mX+bo
regr = linear_model.LinearRegression()

# Entrenamos nuestro modelo de regresion lineal, con la siguiente función
regr.fit(X_train, y_train)

# Hacemos las predicciones segun el modelo de regresion lineal
y_pred = regr.predict(X_train)

# Ahora imprimimos los resultados obtenidos
# Vemos el valor de la pendiente, osea la variable m, el coeficiente de la variable independiente X, en la función lineal
print('Valor de la tangente (m) o Coefficients:=======> %.2f' % regr.coef_)
# Ahora el valor de la constante bo, es decir el valor donde la recta corta el eje Y (Es decir, en X=0)
print('Valor de la constante o Independent term: =====>', regr.intercept_)
# Se imprime el Error Cuadrado Medio
print("Medida dela dispersión delos puntos alrededor de la recta estimada, llamado también,")
print ("Error cuadrado medio(Mean squared error):=> %.3f " % mean_squared_error(y_train, y_pred))
# Puntaje o valor de Varianza. El mejor puntaje es un 1.0
print('Precisión del modelo o  Coeficiente de Determinación:=========> %.3f' % r2_score(y_train, y_pred))
```

```python
print ('Función de la recta del modelo de regresión lineal:  Y=',regr.coef_,'X',regr.intercept_ )
```

```python
# Gráfico de dispersion del comportamiento del Area Nacional versus Produccion Nacional
plt.scatter(produccion_df['Area Nacional (ha)'],produccion_df['Produccion Nacional (ton)'])
plt.title('Area Nacional vs Produccion Nacional')
plt.xlabel('Area Nacional (ha)')
plt.ylabel("Produccion Nacional (ton)")

# A continuación se grafica en colo azul, la funcion lineal obtenida a partir del modelo de regresion lineal
plt.plot(X_train[:,0], y_pred, color='blue', linewidth=3)
plt.show()
```
